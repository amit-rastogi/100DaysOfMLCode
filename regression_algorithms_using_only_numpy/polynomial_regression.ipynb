```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_x = []
train_y = []
#load the data its in format x,y.
for line in open('data_poly.csv'):
    x, y = line.split(',')
    #for fitting a quadratic equation
    train_x.append([1, float(x), float(x)*float(x)])
    train_y.append(float(y))

#convert to numpy array
train_x = np.array(train_x)
train_y = np.array(train_y)
```

```python
#visualize the data
plt.scatter(train_x[:,1], train_y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#calculate the weight matrix
#w = np.dot(np.linalg.inv(np.dot(train_x.T, train_x)), np.dot(train_x.T, train_y))
#alternatively w can be calculated using np.linalg.solve as below
w = np.linalg.solve(np.dot(train_x.T, train_x),np.dot(train_x.T, train_y))
```

```python
#calculate the estimated values of y
yhat = np.dot(train_x, w)
```

```python
#plot the estimated points over the given data points
plt.scatter(train_x[:,1], train_y)
plt.plot(sorted(train_x[:,1]), sorted(yhat))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#calculate accuracy using R-square
num = train_y - yhat
den = train_y - train_y.mean()
rsquared = 1 - num.dot(num)/den.dot(den)
print(rsquared)
```
